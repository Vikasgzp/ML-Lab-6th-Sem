```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
print(df.isnull().sum())
```
