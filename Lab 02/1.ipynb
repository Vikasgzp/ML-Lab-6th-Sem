```python
import numpy as np
import pandas as pd
```

```python
iris_df = pd.read_csv("Iris.csv")
```

```python
sepal_length_array = iris_df['sepal_length'].to_numpy()

print("First 10 values of 'sepal_length' NumPy array:")
print(sepal_length_array[:10])
print(f"Shape of the array: {sepal_length_array.shape}")
```

```python
numbers = [5, 2, 8, 1, 9, 4]
print(f"Original list: {numbers}")

# Bubble sort using while loop
length = len(numbers)
has_swapped = True

while has_swapped:
    has_swapped = False
    index = 0
    while index < length - 1:
        if numbers[index] > numbers[index + 1]:
            # Swap elements
            numbers[index], numbers[index + 1] = numbers[index + 1], numbers[index]
            has_swapped = True
        index += 1
    length -= 1  # Optimization: last element is already sorted

print(f"Sorted list using while loop: {numbers}")
```

```python
sorted_sepal_length = np.sort(sepal_length_array)

print("First 10 values of the sorted 'sepal_length' NumPy array:")
print(sorted_sepal_length[:30])
print(f"Shape of the sorted array: {sorted_sepal_length.shape}")
```

```python
import numpy as np
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()

# Create Boolean mask for Petal Length > 4.0 (Column index 2)
mask = iris.data[:, 2] > 4.0

# Filter rows using Boolean indexing
large_petals = iris.data[mask]

print("Filtered data (Petal Length > 4.0):")
print(large_petals)
```

```python
# Given species array
species = np.array(['Setosa', 'Versicolor', 'Virginica', 'Setosa'])

# Encode: Setosa -> 1, Others -> 0
mask = np.where(species == 'Setosa', 1, 0)

print("Encoded Mask:")
print(mask)
```

```python
import numpy as np
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
data = iris.data.copy()   # Create a copy of the Iris array

# Manually insert NaN at index [0, 0]
data[0, 0] = np.nan

# Locate missing value using np.isnan()
missing_mask = np.isnan(data)

print("Location of missing value:")
print(missing_mask)

# Compute mean of the affected column (column 0), ignoring NaN
col_mean = np.nanmean(data[:, 0])

# Replace NaN with column mean
data[missing_mask] = col_mean

print("\nUpdated data (NaN replaced with column mean):")
print(data[:5])  # show first 5 rows for verification
```
